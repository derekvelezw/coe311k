```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return np.sin(x)

# Interval and number of points
a = 0
b = 2 * np.pi
N = 100

# Step size
h = (b - a) / (N - 1)

# x values
x = np.linspace(a, b, N)

# True derivative
true_derivative = np.cos(x)

# Forward Difference Approximation
forward_diff = np.zeros(N)
for i in range(N - 1):
    forward_diff[i] = (f(x[i] + h) - f(x[i])) / h
forward_diff[-1] = np.nan   # undefined at last point

# Backward Difference Approximation
backward_diff = np.zeros(N)
backward_diff[0] = np.nan   # undefined at first point
for i in range(1, N):
    backward_diff[i] = (f(x[i]) - f(x[i] - h)) / h

# Central Difference Approximation
central_diff = np.zeros(N)
central_diff[0] = np.nan    # undefined at first point
central_diff[-1] = np.nan   # undefined at last point
for i in range(1, N - 1):
    central_diff[i] = (f(x[i] + h) - f(x[i] - h)) / (2 * h)

# Plot all 4 on one canvas with 4 subplots
plt.figure(figsize=(12, 10))

# True derivative
plt.subplot(2, 2, 1)
plt.plot(x, true_derivative)
plt.title("True Derivative: cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

# Forward Difference
plt.subplot(2, 2, 2)
plt.plot(x, true_derivative, label="cos(x)")
plt.plot(x, forward_diff, label="Forward Difference")
plt.title("Forward Difference vs cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Backward Difference
plt.subplot(2, 2, 3)
plt.plot(x, true_derivative, label="cos(x)")
plt.plot(x, backward_diff, label="Backward Difference")
plt.title("Backward Difference vs cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Central Difference
plt.subplot(2, 2, 4)
plt.plot(x, true_derivative, label="cos(x)")
plt.plot(x, central_diff, label="Central Difference")
plt.title("Central Difference vs cos(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
