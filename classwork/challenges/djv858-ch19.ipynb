```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(t, y):
    return -1000*y + 3000 - 2000*np.exp(-t)

def y_exact(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)

def euler_forward(f, t0, tf, y0, h, store_every=1):
    """
    Forward Euler method.
    
    Parameters
    ----------
    f : function
        RHS of ODE: y' = f(t, y)
    t0, tf : float
        Initial and final times
    y0 : float
        Initial condition
    h : float
        Step size
    store_every : int
        Only store every k-th point to reduce memory usage
        
    Returns
    -------
    t_vals : np.ndarray
    y_vals : np.ndarray
    """
    n_steps = int(round((tf - t0) / h))
    
    t = t0
    y = y0
    
    t_list = [t]
    y_list = [y]
    
    for n in range(1, n_steps + 1):
        y = y + h * f(t, y)
        t = t0 + n*h
        
        if n % store_every == 0 or n == n_steps:
            t_list.append(t)
            y_list.append(y)
    
    return np.array(t_list), np.array(y_list)

t0 = 0.0
tf = 30.0
y0 = 0.0

h_values = [0.1, 0.01, 0.001, 0.00001, 0.0000001]

# To avoid gigantic memory use, store fewer points for tiny h
store_map = {
    0.1: 1,
    0.01: 1,
    0.001: 1,
    0.00001: 1000,      # store every 1000th point
    0.0000001: 1000000  # store every 1,000,000th point
}

t_exact = np.linspace(t0, tf, 2000)
y_true = y_exact(t_exact)

plt.figure(figsize=(10, 6))
plt.plot(t_exact, y_true, linewidth=2, label='Analytical Solution')

for h in h_values:
    print(f"\nRunning h = {h} ...")
    
    amp = abs(1 - 1000*h)
    if h < 0.002:
        print(f"  Stable for stiff part since |1 - 1000h| = {amp:.6f} < 1")
    else:
        print(f"  UNSTABLE for stiff part since |1 - 1000h| = {amp:.6f} >= 1")
    
    t_num, y_num = euler_forward(f, t0, tf, y0, h, store_every=store_map[h])
    
    final_error = abs(y_num[-1] - y_exact(t_num[-1]))
    print(f"  Final time t = {t_num[-1]:.6f}")
    print(f"  Euler y(tf)   = {y_num[-1]:.10f}")
    print(f"  Exact y(tf)   = {y_exact(t_num[-1]):.10f}")
    print(f"  Final abs err = {final_error:.10e}")
    
    plt.plot(t_num, y_num, marker='o', markersize=2, linewidth=1, label=f'Euler h={h}')

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title("Forward Euler vs Analytical Solution")
plt.legend()
plt.grid(True)
plt.show()
```
