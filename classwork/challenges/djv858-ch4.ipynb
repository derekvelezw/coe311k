```python
#Leibniz Series for Pi
import numpy as np
import math

true_pi = math.pi

sum = 0.0

for k in range(5000):
    term = ((-1.0)**k)/(2.0*k + 1.0)
    sum += term
    estimated_pi = 4.0 * sum
    
    n_terms = k+1
    if n_terms % 100 == 0:
        rel_error = abs(estimated_pi - true_pi) / abs(true_pi)
        print(f"terms={n_terms:4d}  estimated_pi={estimated_pi:.12f}  rel_error={rel_error:.3e}")
```
