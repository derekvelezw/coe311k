```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

x_scatter = np.random.normal(loc=0.0, scale=1.0, size=80)
y_scatter = np.random.normal(loc=0.0, scale=1.0, size=80)

categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(3, 15, size=len(categories))

hist_data = np.random.normal(loc=70, scale=10, size=300)

fig, axs = plt.subplots(2, 2, figsize=(10, 7))
fig.suptitle("CH1 Matplotlib Freeplay â€” 4 Subplot Types", fontsize=14)

#Line plot
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Plot: sin(x)")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("sin(x)")
axs[0, 0].grid(True)

#Scatter plot
axs[0, 1].scatter(x_scatter, y_scatter)
axs[0, 1].set_title("Scatter Plot")
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("y")

#Bar chart
axs[1, 0].bar(categories, values)
axs[1, 0].set_title("Bar Chart")
axs[1, 0].set_xlabel("Category")
axs[1, 0].set_ylabel("Value")

#Histogram
axs[1, 1].hist(hist_data, bins=20)
axs[1, 1].set_title("Histogram")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Frequency")

plt.plot(x,y)
plt.tight_layout()
plt.show()

fig.savefig("CH1_subplots.png", dpi=200, bbox_inches="tight")
```
