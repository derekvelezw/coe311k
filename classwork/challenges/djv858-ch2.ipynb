```python
def decimal_to_binary(n):

    if n == 0:
        return "0"

    remainders = []

    while n > 0:
        remainders.append(n % 2)
        n = n // 2

    remainders.reverse()
    return ''.join(map(str, remainders))


num = int(input("Enter base-10 number: "))
print("Binary:", decimal_to_binary(num))
```
